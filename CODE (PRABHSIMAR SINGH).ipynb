```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

# Define available stocks
stocks = ["ADANIPORTS.NS", "ASIANPAINT.NS", "AXISBANK.NS", "BAJAJ-AUTO.NS", "BAJAJFINSV.NS",
          "BAJFINANCE.NS", "BHARTIARTL.NS", "BPCL.NS", "BRITANNIA.NS", "CIPLA.NS",
          "COALINDIA.NS", "DIVISLAB.NS", "DRREDDY.NS", "EICHERMOT.NS", "GRASIM.NS",
          "HCLTECH.NS", "HDFCLIFE.NS", "HEROMOTOCO.NS",
          "HINDALCO.NS", "HINDUNILVR.NS", "ICICIBANK.NS", "INDUSINDBK.NS", "INFY.NS",
          "IOC.NS", "ITC.NS", "JSWSTEEL.NS", "KOTAKBANK.NS", "LT.NS", "M&M.NS", 
          "MARUTI.NS", "NTPC.NS", "NESTLEIND.NS", "ONGC.NS", "POWERGRID.NS", "RELIANCE.NS",
          "SBIN.NS", "SUNPHARMA.NS", "TCS.NS", "TATACONSUM.NS", "TATAMOTORS.NS", 
          "TATASTEEL.NS", "TECHM.NS", "TITAN.NS", "ULTRACEMCO.NS", "UPL.NS", "WIPRO.NS"]

# (Other functions remain unchanged)

# Step 3: Define Trading Strategy with Bollinger Bands
def apply_strategy(data, investment_type, initial_investment):
    cash = initial_investment
    shares = 0
    trade_log = []
    portfolio_values = []

    # Define strategy parameters based on investment type
    params = {
        "aggressive": {"buy_threshold": -0.1, "sell_threshold": 0.1, "position_size": 0.2},
        "moderate": {"buy_threshold": -0.07, "sell_threshold": 0.07, "position_size": 0.15},
        "passive": {"buy_threshold": -0.05, "sell_threshold": 0.05, "position_size": 0.1},
    }
    
    # Loop through each date in the historical data
    for date, row in data.iterrows():
        price = row['Close']
        
        # Calculate % deviation from the Bollinger Bands
        deviation = (price - row['20_day_MA']) / row['20_day_MA']
        
        # Trigger buy/sell based on thresholds
        if deviation <= params[investment_type]["buy_threshold"] and cash > 0:
            # Buy action
            buy_amount = cash * params[investment_type]["position_size"]
            shares_bought = buy_amount // price
            cash -= shares_bought * price
            shares += shares_bought
            trade_log.append({"type": "buy", "date": date, "price": price, "shares": shares_bought})
        
        elif deviation >= params[investment_type]["sell_threshold"] and shares > 0:
            # Sell action
            sell_amount = shares * price
            cash += sell_amount
            trade_log.append({"type": "sell", "date": date, "price": price, "shares": shares})
            shares = 0

        # Track portfolio value for metrics
        portfolio_value = cash + shares * price
        portfolio_values.append(portfolio_value)

    # Calculate Metrics
    returns = np.diff(portfolio_values) / portfolio_values[:-1]
    final_value = portfolio_values[-1]
    roi = (final_value - initial_investment) / initial_investment * 100
    sharpe_ratio = np.mean(returns) / np.std(returns) * np.sqrt(252)
    max_drawdown = (np.min(portfolio_values) / np.max(portfolio_values)) - 1
    win_rate = len([trade for trade in trade_log if trade['type'] == "sell" and trade['price'] > trade_log[0]['price']]) / len(trade_log)
    volatility = np.std(returns) * np.sqrt(252)
    profit_factor = sum([trade['price'] for trade in trade_log if trade['type'] == "sell"]) / \
                    sum([trade['price'] for trade in trade_log if trade['type'] == "buy"])

    # Compute drawdown
    drawdowns = (portfolio_values - np.maximum.accumulate(portfolio_values)) / np.maximum.accumulate(portfolio_values)

    metrics = {
        "ROI": roi,
        "Sharpe Ratio": sharpe_ratio,
        "Max Drawdown": max_drawdown,
        "Win Rate": win_rate,
        "Volatility": volatility,
        "Profit Factor": profit_factor
    }

    return final_value, trade_log, portfolio_values, metrics, returns, drawdowns

# Step 4: Run Simulation for Each Investment Type and Compare
def run_simulation(stock, investment, start_date, end_date, investment_type):
    data = load_data(stock, start_date, end_date)
    
    if investment_type == "all":
        results = {}
        for mode in ["aggressive", "moderate", "passive"]:
            final_value, trade_log, portfolio_values, metrics, returns, drawdowns = apply_strategy(data, mode, investment)
            results[mode] = {"final_value": final_value, "trade_log": trade_log, "portfolio_values": portfolio_values, 
                             "metrics": metrics, "returns": returns, "drawdowns": drawdowns}
    else:
        final_value, trade_log, portfolio_values, metrics, returns, drawdowns = apply_strategy(data, investment_type, investment)
        results = {investment_type: {"final_value": final_value, "trade_log": trade_log, "portfolio_values": portfolio_values, 
                                     "metrics": metrics, "returns": returns, "drawdowns": drawdowns}}

    return results

# Step 5: Display Results and Compare
def display_results(results, initial_investment):
    for mode, result in results.items():
        final_value = result["final_value"]
        trade_log = result["trade_log"]
        metrics = result["metrics"]
        returns = result["returns"]
        drawdowns = result["drawdowns"]

        print(f"\nInvestment Mode: {mode.capitalize()}")
        print(f"Initial Investment: ${initial_investment}")
        print(f"Final Portfolio Value: ${final_value:.2f}")
        print(f"Return: {(final_value - initial_investment) / initial_investment * 100:.2f}%")
        print(f"Total Trades: {len(trade_log)}")
        for trade in trade_log:
            print(f"{trade['type'].capitalize()} on {trade['date']} at ${trade['price']:.2f} for {trade['shares']} shares")

        print("\nMetrics:")
        for metric, value in metrics.items():
            print(f"{metric}: {value:.2f}")

        # Plot equity curve
        plt.figure(figsize=(14, 5))
        plt.plot(result["portfolio_values"], label="Equity Curve")
        plt.xlabel("Time")
        plt.ylabel("Portfolio Value")
        plt.title(f"Equity Curve for {mode.capitalize()} Investment")
        plt.legend()
        plt.show()

        # Plot distribution of returns
        plt.figure(figsize=(10, 5))
        plt.hist(returns, bins=50, alpha=0.7, color='blue')
        plt.title(f"Distribution of Returns for {mode.capitalize()} Investment")
        plt.xlabel("Return")
        plt.ylabel("Frequency")
        plt.show()

        # Plot drawdown chart
        plt.figure(figsize=(14, 5))
        plt.plot(drawdowns, color='red', label="Drawdown")
        plt.fill_between(range(len(drawdowns)), drawdowns, color='red', alpha=0.3)
        plt.title(f"Drawdown Chart for {mode.capitalize()} Investment")
        plt.xlabel("Time")
        plt.ylabel("Drawdown")
        plt.legend()
        plt.show()

# Main Function
def main():
    stock, investment, start_date, end_date, investment_type = user_input_parameters()
    results = run_simulation(stock, investment, start_date, end_date, investment_type)
    display_results(results, investment)

# Run the simulation
main()
```
